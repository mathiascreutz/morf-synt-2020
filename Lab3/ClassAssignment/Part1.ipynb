To install the software:

1. Log in to your account on a university Windows desktop computer.
2. Click on the University of Helsinki "flame logo" icon, which is in the "tray" at the bottom right corner of the screen.
3. A menu should appear. From the menu choose "Software Center".

6. Install this package.
7. Installation may take 10 minutes or more.
8. Wait or do something else until the installation is complete.

When installation is complete, you are ready to start the first demo. Perform the following steps:

1. Click on the magnifying glass or circle in the lower left corner of the screen.
2. Enter the word IDLE in the text field.
3. Wait for Windows to suggest some match.
4. There are probably multiple choices: you need to pick IDLE with a version number starting in 3, such as 3.4, and _additionally_ there has to be a mention that it is a 32-bit version.
5. If no such option appears, and you haven't made any mistake yourself, then the necessary software might not have installed successfully on your account. It is also possible that you have to repeat your search for IDLE after a little while, and it will show then. Or, if this does not help, you can join someone else who has the software available.
6. Click on the IDLE 3.x 32-bit version and wait for the IDLE window to appear.
7. In the IDLE window, first type `import nltk` and hit enter. Expect there to be no error messages at this point.
8. Next, type `nltk.app.rdparser()`. Now a warning message in red might be displayed, but you should also see a new window appear on your screen. It is the first demo and should look something like this:



### Recursive Descent Parser Demo

Play around with the Recursive Descent parser demo. Make sure that you understand how the parsing takes place step by step. This is a simulation of an algorithm that the computer executes as a computer program in order to parse a sentence according to a given context free grammar (CFG). On the left hand side you see the context free grammar. At the bottom on the right hand side you see an example sentence that is being parsed.

Try to use only the buttons **Expand** and **Match**, and if necessary **Backtrack**. Preferably, do not use the _cheat buttons_ **Step** and **Autostep**. The cheat buttons know which next step is the correct one, but you should test for yourself what happens when you expand, match or backtrack. Do you understand the meaning of these steps?

Note that you can *edit the grammar or text yourself*! You find the edit functionality under the Edit menu of the demo window.

### Shift-Reduce Parser Demo

Next, close the recursive descent parser demo window. Go back to the IDLE window and type `nltk.app.srparser()`. A new demo window should appear. This parser works differently than the recursive descent parser. Now, there are two other operations, called **Shift** and **Reduce**.

Again, try not to cheat using the **Step** button. If necessary, use the **Undo** button (which, however, does not exist in the actual shift-reduce algorithm).

Do you understand the meaning of the operations Shift and Reduce? In what fundamental way does the shift-reduce parser differ from the recursive descent parser? Which one do you prefer?

Also in this demo you can modify the grammar or text in the Edit menu.

When you are done, close the demo window, and continue to Part 2. Part 2 works on every computer, so you can go back to your own computer if you have joined forces with someone else.
